```python
import random
import pandas as pd 
import numpy as np
import  matplotlib.pyplot as plt 
age=[]
for i in range (0,100):
    age.append(random.randint(1,100))
```

```python
type(age)
```

```python
# sd=(x-mean)/st
```

```python
sd=[]
m=np.mean(age)
st=np.std(age)
```

```python
for x in age:
    sd.append((x-m)/st)
```

```python
print(sd)
```

```python
np.mean(sd)
```

```python
plt.scatter(range(0,100),age)
```
